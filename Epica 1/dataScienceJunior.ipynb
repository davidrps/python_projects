Agile/Scrum es un enfoque de gestión de proyectos que se centra en la entrega continua de valor y la flexibilidad ante cambios. El proceso Agile es "iterativo" lo que significa que se repite varias veces con el objetivo de mejorar el producto.

Una **"épica"** es una historia de usuario de gran alcance que se divide en tareas más pequeñas.

Una **"historia de usuario"** es una descripción de un problema o necesidad de un usuario.

Una **"card"** o **"tarjeta"** es una herramienta utilizada para describir y asignar tareas en un proyecto Scrum.

Una **"conversación"** es una discusión en equipo para aclarar detalles o resolver dudas sobre una tarea o historia de usuario.

Una **"confirmación"** es el proceso de verificar que una tarea o historia de usuario ha sido completada satisfactoriamente.

Un **"sprint"** es un período de tiempo (generalmente de 2 a 4 semanas) durante el cual se desarrolla un conjunto específico de tareas o historias de usuario.

El **"backlog"** es una lista de tareas o historias de usuario pendientes.

El **"kickOff"** o **"presentación de nuevo proyecto"** es una reunión en la que se presenta el proyecto y se establecen los objetivos.

La **"stand-up meeting"** es una reunión diaria en la que los miembros del equipo informan sobre el progreso de las tareas.

La **"hoja de ruta"** es un plan detallado para alcanzar los objetivos a largo plazo del proyecto.

Los **"skills"** son habilidades o destrezas que una persona posee. En el contexto de una carrera profesional, las skills son las habilidades y conocimientos específicos que una persona tiene en un área determinada. Algunos ejemplos de skills en el ámbito laboral pueden incluir habilidades técnicas como programación, habilidades de liderazgo, habilidades de comunicación, habilidades analíticas, habilidades en el manejo de herramientas específicas, entre otros. En el contexto de la búsqueda de empleo, las habilidades o skills son un factor importante a considerar tanto para el candidato como para el empleador, ya que son un indicador de la capacidad de una persona para desempeñar un trabajo específico.

Un **"stack de programación"** es un conjunto de herramientas, tecnologías y lenguajes de programación utilizados para desarrollar una aplicación o sistema. Un stack de programación esta compuesto por tres capas:

- Capa de Front-End: Es la parte visible de la aplicación que interactúa con el usuario. Se encarga de la interfaz gráfica de usuario (GUI) y suele ser desarrollada con lenguajes como HTML, CSS y JavaScript.

- Capa de Back-End: Es la parte invisible de la aplicación que se encarga de manejar los datos y la lógica de negocio. Se suele desarrollar con lenguajes como Python, Java, Ruby o C#.

- Capa de Base de datos: Es la parte de la aplicación encargada de almacenar y recuperar los datos. Se suelen usar bases de datos relacionales como MySQL, PostgreSQL, y bases de datos NoSQL como MongoDB, Cassandra.

Existen varios stacks de programación populares. Algunos ejemplos son:

- Stack LAMP (Linux, Apache, MySQL, PHP/Perl/Python): utilizado para desarrollar aplicaciones web en entorno Linux.
- Stack MEAN (MongoDB, Express.js, AngularJS, Node.js): utilizado para desarrollar aplicaciones web full-stack con JavaScript.
- Stack MERN (MongoDB, Express.js, React, Node.js): utilizado para desarrollar aplicaciones web full-stack con JavaScript.
- Stack Ruby on Rails (Ruby, Ruby on Rails, PostgreSQL): utilizado para desarrollar aplicaciones web en entorno Ruby.

Cada stack tiene sus propias ventajas y desventajas y se debe elegir en función del proyecto y de las necesidades específicas del equipo de desarrollo.

## Junior/Senior

¿Qué significa **Junior** en un puesto de trabajo?
El empleado Junior tiene una experiencia laboral menor a 2 años. Esto significa que aún sigue adquiriendo conocimientos del puesto y suele trabajar con tareas sencillas y supervisadas.

¿Qué es ser **Senior**?
Un trabajador se considera senior a partir de los cinco o seis años de experiencia. Por ello, habitualmente, se relaciona a este tipo de perfiles con una edad avanzada.

## ¿Qué es Jira?

Jira es una herramienta de gestión de proyectos y seguimiento de errores desarrollada por Atlassian. Es comúnmente utilizada en proyectos de software, pero también se utiliza en proyectos de otras áreas. Jira permite a los equipos crear, asignar y seguir tareas, bugs y historias de usuario, así como planificar sprints y releases.

## ¿Cómo se crean cards en un calendario?

Para crear una tarjeta o card en Jira, se pueden seguir los siguientes pasos:

- Iniciar sesión en Jira.

- Seleccionar el proyecto en el que se quiere crear la tarjeta.

- Hacer clic en el botón "Create" (Crear) en la barra de herramientas.

- Seleccionar el tipo de tarea que se quiere crear (ejemplo: tarea, bug, historia de usuario)

- Rellenar los detalles de la tarea en el formulario, como el título, la descripción, el asignado, el tipo, el estado, entre otros.

- Hacer clic en "Create" (crear) para crear la tarjeta.

Una vez creada la tarjeta se puede asignar a un miembro del equipo, establecer una fecha de vencimiento, agregar comentarios, entre otras acciones. Para ver las tarjetas en un calendario, se puede utilizar el plugin de calendario de Jira, el cual permite visualizar las tareas en un calendario y así tener una mejor planificación de los plazos.

## Términos importantes en Data Science

- **Data mining**: proceso de descubrimiento de patrones y conocimiento a partir de grandes conjuntos de datos.

- **Machine learning**: técnica que permite que las computadoras aprendan de los datos sin ser explícitamente programadas.

- **Modelado estadístico**: proceso de construir un modelo matemático para entender la relación entre variables y hacer predicciones.

- **Big data**: grandes conjuntos de datos que superan la capacidad de procesamiento de una sola computadora.

- **Algoritmo**: conjunto de reglas y procedimientos para resolver un problema o realizar una tarea específica.

- **Visualización de datos**: representación gráfica de datos para facilitar la comprensión y el análisis.

- **Aprendizaje automático**: técnica para que las computadoras aprendan por sí mismas a partir de los datos.

- **Regresión**: técnica estadística utilizada para entender la relación entre variables y hacer predicciones.

- **Clasificación**: técnica utilizada para asignar una etiqueta o categoría a una observación de datos.

- **Análisis de componentes principales (PCA)**: técnica utilizada para simplificar y reducir la dimensionalidad de los datos.

## Recomendaciones para conseguir entrevistas de trabajo en Data Science

- Amplía tus conocimientos en ciencia de datos: Asegúrate de estar al día con las últimas tendencias y herramientas en el campo. Esto puede incluir aprender sobre el procesamiento de grandes volúmenes de datos, aprendizaje automático y visualización de datos.

- Haz un buen portafolio: Muestra tu experiencia y habilidades mediante proyectos y ejemplos concretos de tus trabajos anteriores. Asegúrate de incluir una descripción detallada de tu rol en cada proyecto y los resultados obtenidos.

- Participa en comunidades y eventos de ciencia de datos: Únete a grupos de discusión y asiste a conferencias y eventos relacionados con la ciencia de datos. Esto te ayudará a mantenerte actualizado y conectado con otros profesionales en el campo.

- Crea una buena red de contactos: Trata de conectarte con profesionales en la industria y en la academia. Esto te ayudará a conocer sobre oportunidades de trabajo y te brindará apoyo en tu búsqueda.

- Asegúrate de tener un buen currículum y carta de presentación: Asegúrate de que tu currículum sea conciso, claro y llame la atención. Asegúrate de que tu carta de presentación esté personalizada para la empresa a la que estás aplicando y destaque tus habilidades y experiencias relevantes.

- Aplica a muchas ofertas de trabajo: No te rindas si no recibes una respuesta positiva en tu primera entrevista, sigue buscando oportunidades y aplicando a ellas.

- No subestimes la importancia de las habilidades blandas: Es importante tener una buena comunicación, capacidad para trabajar en equipo y habilidades para presentar y explicar tus resultados.
